```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

# Download required NLTK resources
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('punkt', force=True)  # Force re-download the tokenizer
```

```python


# Load dataset (Replace with actual filename)
df = pd.read_csv('climate_nasa.csv')



```

```python
# Strip spaces from column names (if any)
df.columns = df.columns.str.strip()

# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['date'], errors='coerce')

# Convert LikesCount & CommentsCount to numeric (if not already)
df['LikesCount'] = pd.to_numeric(df['likesCount'], errors='coerce')
df['CommentsCount'] = pd.to_numeric(df['commentsCount'], errors='coerce')

# Initialize NLP tools
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))
```

```python
def clean_text(text):
    if not isinstance(text, str):  # Ensure input is a string
        return ""

    text = text.lower()
    text = re.sub(r'http\S+|www.\S+', '', text)  # Remove URLs
    text = re.sub(r'[^a-z\s]', '', text)  # Remove special characters
    words = text.split()  # Use split() instead of word_tokenize()
    words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words]  
    return ' '.join(words)


print(clean_text("This is a test comment!"))  # If this gives an error, the function is incorrect

```

```python
# Apply text cleaning function
df['Cleaned_Text'] = df['text'].apply(clean_text)  

# Save cleaned dataset
df.to_csv('cleaned_dataset.csv', index=False)

# Display first few rows to verify
print(df[['text', 'Cleaned_Text']].head(10))
```

```python

from nltk.sentiment import SentimentIntensityAnalyzer

# Download VADER Lexicon (Run this once)
nltk.download('vader_lexicon')

# Initialize the Sentiment Intensity Analyzer
sia = SentimentIntensityAnalyzer()
```

```python
# Function to get sentiment score
def get_sentiment(text):
    if not isinstance(text, str) or text.strip() == "":
        return 0  # Neutral score for empty/missing text
    return sia.polarity_scores(text)['compound']

# Apply sentiment analysis
df['Sentiment_Score'] = df['Cleaned_Text'].apply(get_sentiment)

# Assign sentiment labels based on score
df['Sentiment'] = df['Sentiment_Score'].apply(lambda score: 'Positive' if score > 0.05 
                                              else ('Negative' if score < -0.05 else 'Neutral'))

# Show some results
print(df[['Cleaned_Text', 'Sentiment_Score', 'Sentiment']].head(10))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Count sentiment labels
plt.figure(figsize=(8,5))
sns.countplot(x=df['Sentiment'], palette="coolwarm")
plt.title("Sentiment Distribution")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])  # Ensure datetime format
df['Year-Month'] = df['Date'].dt.to_period('M')  # Group by month

# Compute average sentiment per month
sentiment_trend = df.groupby('Year-Month')['Sentiment_Score'].mean()

# Plot sentiment trend
plt.figure(figsize=(12,5))
plt.plot(sentiment_trend.index.astype(str), sentiment_trend.values, marker='o', linestyle='-')
plt.xticks(rotation=45)
plt.title("Sentiment Trend Over Time")
plt.xlabel("Time (Year-Month)")
plt.ylabel("Average Sentiment Score")
plt.grid()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set figure size
plt.figure(figsize=(12, 5))

# Boxplot for LikesCount per Sentiment
sns.boxplot(x='Sentiment', y='LikesCount', data=df, palette="coolwarm")
plt.title("Likes Distribution by Sentiment")
plt.xlabel("Sentiment")
plt.ylabel("Number of Likes")
plt.ylim(0, df['LikesCount'].quantile(0.95))  # Remove extreme outliers
plt.show()

# Boxplot for CommentsCount per Sentiment
plt.figure(figsize=(12, 5))
sns.boxplot(x='Sentiment', y='CommentsCount', data=df, palette="coolwarm")
plt.title("Comments Distribution by Sentiment")
plt.xlabel("Sentiment")
plt.ylabel("Number of Comments")
plt.ylim(0, df['CommentsCount'].quantile(0.95))  # Remove extreme outliers
plt.show()
```

```python
# Convert Sentiment labels to numeric values for correlation
sentiment_mapping = {'Positive': 1, 'Neutral': 0, 'Negative': -1}
df['Sentiment_Numeric'] = df['Sentiment'].map(sentiment_mapping)

# Check correlation between sentiment & engagement
correlation_matrix = df[['Sentiment_Numeric', 'LikesCount', 'CommentsCount']].corr()
print(correlation_matrix)

# Heatmap for better visualization
plt.figure(figsize=(8,5))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Between Sentiment & Engagement")
plt.show()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
import numpy as np
```

```python
# Vectorize text data using CountVectorizer (Bag of Words)
vectorizer = CountVectorizer(max_df=0.9, min_df=5, stop_words='english', max_features=1000)
X = vectorizer.fit_transform(df['Cleaned_Text'])
```

```python
# Define number of topics
num_topics = 3  # Adjust this based on results

# Train LDA model
lda = LatentDirichletAllocation(n_components=num_topics, random_state=42)
lda.fit(X)
```

```python
# Get words from vectorizer
words = np.array(vectorizer.get_feature_names_out())

# Show top words for each topic
for topic_idx, topic in enumerate(lda.components_):
    print(f"🔹 Topic {topic_idx+1}: ", [words[i] for i in topic.argsort()[-10:]])
```

```python
# Select key columns
df_final = df[['Date', 'Cleaned_Text', 'Sentiment', 'Sentiment_Score', 'LikesCount', 'CommentsCount']]

# Save to CSV
df_final.to_csv('final_analysis_results.csv', index=False)

print(" Final results saved as 'final_analysis_results.csv'")
```
